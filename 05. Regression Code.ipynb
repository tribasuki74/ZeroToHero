```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

based = ['01', '02', '03', '04', '05', '06']
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([15, 11, 2, 8, 25, 32])

print(x)
print(y)
print()


# LinearRegression
model = LinearRegression().fit(x, y)

r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
print()

y_pred = model.predict(x)
print(f"predicted response:\n{y_pred}")

plt.plot(based, y)
plt.plot(based, y_pred)
plt.show()
```

```python
# Multiple Linear Regression

x = [[5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34]]
y = np.array([15, 11, 2, 18, 25, 32])

x, y = np.array(x), np.array(y)
print(x)
print(y)
print()

model = LinearRegression().fit(x, y)

r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
print()

y_pred = model.predict(x)
print(f"predicted response:\n{y_pred}")

based = ['01', '02', '03', '04', '05', '06']

plt.plot(based, y)
plt.plot(based, y_pred)
plt.show()
```

```python
# Polynomial Regression
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x = [[5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34]]
y = np.array([15, 11, 2, 18, 25, 32])

# x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
# y = np.array([5, 20, 14, 32, 22, 38])

x_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x)
# x_ = PolynomialFeatures(degree=2, include_bias=True).fit_transform(x)
print(x)
print(x_)
print(y)
print()

model = LinearRegression().fit(x_, y)

r_sq = model.score(x_, y)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
print()

y_pred = model.predict(x_)
print(f"predicted response:\n{y_pred}")

based = ['01', '02', '03', '04', '05', '06']

plt.plot(based, y)
plt.plot(based, y_pred)

plt.show()
```

```python
import numpy as np
from scipy.stats import linregress
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# input data
y = np.array([76.6, 118.6, 200.8, 362.3, 648.9])
x = np.array([10, 20, 40, 80, 160])

# scipy linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_pred = intercept + slope * x

mse = mean_squared_error(y_true=y, y_pred=y_pred)
rmse = mean_squared_error(y_true=y, y_pred=y_pred)

print('scipy intercept: {:.6f}'.format(intercept))
print('scipy slope: {:.6f}'.format(slope))
print('scipy MSE: {:.6f}'.format(mse))
print('scipy RMSE: {:.6f}'.format(rmse))

# sklearn linear regression
reg = LinearRegression().fit(x.reshape(- 1, 1), y)
y_pred = reg.predict(x.reshape(- 1, 1))

mse = mean_squared_error(y_true=y, y_pred=y_pred)
rmse = mean_squared_error(y_true=y, y_pred=y_pred)

print('sklearn intercept: {:.6f}'.format(reg.intercept_))
print('sklearn slope: {:.6f}'.format(reg.coef_[0]))
print('sklearn MSE: {:.6f}'.format(mse))
print('sklearn RMSE: {:.6f}'.format(rmse))
```
