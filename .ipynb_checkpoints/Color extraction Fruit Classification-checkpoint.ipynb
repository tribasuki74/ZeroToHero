```python
# train_path = "/content/gdrive/My Drive/FruitClassification/dataset/train/train/"
# test_path = "/content/gdrive/My Drive/FruitClassification/dataset/test/test/"

train_path = "https://drive.google.com/drive/folders/1ylQ61KFu5VsDK_ETm-S3u8dKrRD7DqrW?usp=sharing"
test_path = "https://drive.google.com/drive/folders/1B3u5yvYQnMJbhMJRbKtH5Bgh95xzTAeM?usp=sharing"
```

```python
data_dir = pathlib.Path(train_path)
class_names = np.array(sorted([item.name for item in data_dir.glob('*')]))
print(class_names)
```

```python
# Plot a grid of images from a training dataset

def view_random_image(target_dir, target_class):
  # Setup target directory (we'll view images from here)
  target_folder = target_dir+target_class

  # Get a random image path
  random_image = random.sample(os.listdir(target_folder), 100)

  # Read in the image and plot it using matplotlib
  img = mpimg.imread(target_folder + "/" + random_image[0])
  plt.imshow(img)
  plt.title(target_class)
  plt.axis("off");

  return img
```

```python
train_val_data = {'path' : [],
       'filename': [],
       'label': []}
for dirpath, dirnames, filenames in os.walk(train_path):
    for f in filenames:
        train_val_data['path'].append(dirpath)
        train_val_data['filename'].append(f)
        train_val_data['label'].append(f.split('_')[0])
```

```python
train_val_data_df = pd.DataFrame(train_val_data)
# train_val_data_df = train_val_data_df.sample(frac=0.1)
display(train_val_data_df.head())
print(train_val_data_df.shape)
```

```python
images = []
label = []

idx = 0
total = train_val_data_df.shape[0]
for _, d in train_val_data_df.iterrows():
    idx += 1
    if(idx % 250 == 0):
        print('{:.0f} images was read from {:.0f}'.format(idx, total))

    imgpath = os.path.join(d['path'],d['filename'])
    image = imread(imgpath)
    images.append(image)
    label.append(d['label'])

images = np.array(images)
labels = np.array(label)
print(f"Complete data images shape: {images.shape} and label shape: {labels.shape}")
```

```python
from google.colab.patches import cv2_imshow

f, axarr = plt.subplots(2,5, figsize=(15, 5))
axarr[0,0].imshow(images[0])
axarr[0,1].imshow(images[1])
axarr[0,2].imshow(images[2])
axarr[0,3].imshow(images[3])
axarr[0,4].imshow(images[4])
axarr[1,0].imshow(images[5])
axarr[1,1].imshow(images[6])
axarr[1,2].imshow(images[7])
axarr[1,3].imshow(images[8])
axarr[1,4].imshow(images[9])


print(images[0].shape)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

class_indices = dict(zip(class_names, range(len(class_names))))
labels_encoded = list(map(class_indices.get, labels))
label_categorical = tf.keras.utils.to_categorical(labels_encoded, num_classes=len(class_names))

enc = pd.DataFrame({'labels': labels})
label_names = enc['labels'].unique()
label_names.sort()

enc['labels']= label_encoder.fit_transform(enc['labels'])

result = {key: val for key, val in zip(label_names, range(len(label_names)))}
display(result)
label_counts = enc['labels'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
plt.bar(label_names[label_counts.index], label_counts.values, color='skyblue')

plt.xticks(rotation='vertical')
plt.xlabel('Label Name')
plt.ylabel('Frequency')
plt.title('Label Frequencies')
plt.legend(['Labels'])
plt.show()
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(images, label_categorical, test_size = 0.4, random_state=101)

print("X train data : ", len(X_train))
print("X label data : ", len(X_val))
print("Y test data : ", len(Y_train))
print("Y label data : ", len(Y_val))
```

```python
print ("train data shape after the split: ", X_train.shape)
print ('new validation data shape: ', X_val.shape)
print ("validation labels shape: ", Y_val.shape)
```

```python
datagen = ImageDataGenerator(horizontal_flip=False,
                             vertical_flip=False,
                             rotation_range=0,
                             zoom_range=0.2,
                             width_shift_range=0,
                             height_shift_range=0,
                             shear_range=0,
                             fill_mode="nearest")
```

```python
pretrained_model = tf.keras.applications.DenseNet121(input_shape=(100,100,3),
                                                      include_top=False,
                                                      weights='imagenet',
                                                      pooling='avg')
pretrained_model.trainable = False
```

```python
def create_custom_cnn(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        MaxPooling2D(2, 2),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Flatten(),
        Dropout(0.5),
        Dense(512, activation='relu'),
        Dropout(0.2),
        Dense(num_classes, activation='softmax')
    ])
    return model

# Define the model parameters
input_shape = (100, 100, 3)  # Adjust based on your dataset preprocessing
num_classes = 33  # Adjust based on the number of fruit classes in your dataset

# Create the CNN model
model = create_custom_cnn(input_shape, num_classes)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['acc'])
```

```python
# Assuming X_train and Y_train are already defined and preprocessed
history = model.fit(
    datagen.flow(X_train, Y_train, batch_size=64),
    validation_data=(X_val, Y_val),
    epochs=5
)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = model.predict(X_val)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(Y_val, axis=1)

# conf_matrix = confusion_matrix(y_true, y_pred_classes)
# print("Confusion Matrix:\n", conf_matrix)
cm = confusion_matrix(y_true, y_pred_classes)

# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Create a heatmap using seaborn
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=class_names,
            yticklabels=class_names)

# Add labels and title
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('Confusion Matrix')

# Show the plot
plt.show()


report = classification_report(y_true, y_pred_classes, target_names=[f'{class_names[i]}' for i in range(len(class_names))])
print("\nClassification Report:\n", report)
```

```python
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend(loc=0)
plt.figure()

plt.show()
```

```python
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend(loc=0)
plt.figure()

plt.show()
```
